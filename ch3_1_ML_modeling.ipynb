```python
# Hold out 방법

import pandas as pd
import numpy as np
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_test.csv')
```

```python
train_x = train.drop(columns='grade')
train_y = train['grade']
test_x = test.drop(columns='grade')
test_y = test['grade']
```

```python
from sklearn.model_selection import train_test_split
train_x_sub, valid_x, train_y_sub, valid_y = train_test_split(train_x, train_y, test_size=0.3, random_state=42)
```

```python
print(train_x_sub.shape)
print(valid_x.shape)
print(train_y_sub.shape)
print(valid_y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_x_sub, train_y_sub)
```

```python
pip install -U scikit-learn
```

```python
from sklearn.metrics import mean_squared_error
pred_val = lr.predict(valid_x)
mse = mean_squared_error(valid_y, pred_val)
rmse = np.sqrt(mse)
print(rmse)
```

```python
#k-fold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

cv_score = cross_val_score(lr, train_x, train_y, scoring = 'neg_mean_squared_error')

rmse_score = np.sqrt(-cv_score)
mean_rmse_score = np.mean(rmse_score)
```

```python
print(rmse_score)
print(mean_rmse_score)
```

?? 이상하다 hold out rmse가 더 높아야 하는데 kfold rmse가 더 높게 나옴

```python
#suffle 옵션
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_score2 = cross_val_score(lr, train_x, train_y, scoring = 'neg_mean_squared_error', cv=cv)
rmse_score2 = np.sqrt(-cv_score2)
```

```python
# Grid Search CV
import pandas as pd
import numpy as np


train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_test.csv')

train_x = train.drop(columns='grade')
train_y = train['grade']
test_x = test.drop(columns='grade')
test_y = test['grade']

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state = 1)
```

```python
params = rf.get_params()

for param_name, param_value in params.items():
  print(param_name ,param_value)
```

```python
param_grid = {
    'max_depth' : [10, 20, 30],
    'ccp_alpha' : [0.1, 0.3, 0.5]  }

rf_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 5, scoring = 'neg_mean_squared_error')
rf_search.fit(train_x, train_y)
```

```python
best_params = rf_search.best_params_
print(best_params)
```

```python
mean_rmse_score3 = np.sqrt(-rf_search.best_score_)
print(mean_rmse_score3)
```

#Pipeline

- 정형 데이터

```python
#data load
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_test.csv')

#x, y 분리
train_x = train.drop(columns = 'grade', axis = 1)
train_y = train['grade']

test_x = test.drop(['grade'], axis = 1)
test_y = test['grade']

#preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR

svr_pipe = Pipeline([
                      ('preprocess', StandardScaler()),
                      ['regressor', SVR()]
])
```

```python
svr_pipe.fit(train_x, train_y)
```

```python
#pipeline

from sklearn.pipeline import make_pipeline
from sklearn.svm import SVR

svr_pipe2 = make_pipeline(
    StandardScaler(),
    SVR()
)
svr_pipe2.fit(train_x, train_y)
```

```python
from sklearn.model_selection import cross_val_score

cv_score4 = cross_val_score(svr_pipe, train_x, train_y, scoring = 'neg_mean_squared_error', cv = 5)
rmse_score4 = np.sqrt(-cv_score4)
mean_rmse_score4 = np.mean(rmse_score4)
```

```python
mean_rmse_score4
```

```python
SVR().get_params()
```

```python
SVR_param = {'regressor__C': np.arange(1,100,20)}

from sklearn.model_selection import GridSearchCV
SVR_search = GridSearchCV(estimator = svr_pipe, param_grid = SVR_param, cv = 5, scoring = 'neg_root_mean_squared_error')
SVR_search.fit(train_x, train_y)
```

```python
print('Best parameter:', SVR_search.best_params_)
print('RMSE:', -SVR_search.best_score_)
```

- 범주형, 수치형 혼합

```python
train = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/data/main/s11_test2.csv')

train_x = train.drop(['grade'], axis = 1)
train_y = train['grade']
test_x = test.drop(['grade'], axis = 1)
test_y = test['grade']
```

```python
num_col = train_x.select_dtypes('number').columns.tolist()
cat_col = train_x.select_dtypes('object').columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

cat_preprocess = make_pipeline(OneHotEncoder(handle_unknown = 'ignore', sparse_output = False))
num_preprocess = make_pipeline(SimpleImputer(strategy = 'mean'), StandardScaler())
```

```python
from sklearn.compose import ColumnTransformer
preprocess = ColumnTransformer(
    [("num", num_preprocess, num_col),
     ("cat", cat_preprocess, cat_col)]
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR

full_pipe = Pipeline(
    [
        ('preprocess', preprocess),
        ("regressor", SVR())
    ]
)
```

```python
SVR_param = {'regressor__C': np.arange(1, 100, 20)}

from sklearn.model_selection import GridSearchCV
SVR_search = GridSearchCV(estimator = full_pipe,
                          param_grid = SVR_param,
                          cv = 5,
                          scoring = 'neg_root_mean_squared_error')
SVR_search.fit(train_x, train_y)

print('best parameter :' ,SVR_search.best_params_)
print('RMSE : ', -SVR_search.best_score_)
```

```python
#최종 예측
test_pred = SVR_search.predict(test_x)
test_pred = pd.DataFrame(test_pred, columns = ['pred'])
```

### lambda 함수 파생변수, 일반 def 파생변수
- lambda 매개변수 : 표현식
- lambda x : **True일 때 값** if **조건** else **else일 때 조건**

누적확진률 컬럼에서 누적확진률의 평균보다 높으면 high 낮으면 low로 변경

```python
corona_df_copy["누적확진율"]
```

```python
#람다함수 사용
corona_df_copy["H/L"] = corona_df_copy["누적확진율"].apply(lambda x : "High" if (x > mean_val) else "Low")
corona_df_copy["H/L"].value_counts()  #조건식이 참일 때 값을 먼저 쓴다
```

```python
## apply() 일반 함수
## 매개변수 값이 누적확진율 평균보다 높으면 High
## 조건식이 거짓이면 Low 

def HL(num):
    if num > mean_val:
        return "High"
    else:
        return "Low"
             

corona_df_copy["누적확진율"].apply(HL)
```

```python
## 요일이라는 파생변수 만들기
## 등록일시 datetime 변경 후 기준으로 요일 출력
# 0부터 6까지ㅟㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅣㅓㅓ

corona_df_copy["등록일시"] = pd.to_datetime(corona_df_copy["등록일시"])
```

```python
corona_df_copy["요일"] = corona_df_copy["등록일시"].dt.weekday
corona_df_copy.head()
```

```python
## apply 함수를 이영해서 숫자로 변경

def change_week(x):
    if x == 0 :
        return "월"
    elif x == 1 :
        return "화"
    elif x == 2 :
        return "수"
    elif x == 3 :
        return "목"
    elif x == 4 :
        return "금"
    elif x == 5 :
        return "토"
    else :
        return "일"

corona_df_copy["요일"] = corona_df_copy["요일"].apply(change_week)
```

```python
corona_df_copy.head()
```

```python
week_corona = corona_df_copy.groupby("요일").mean()
```

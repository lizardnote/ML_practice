## Matplotlib

```python
 #!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
#라인 그래프 출력
plt.plot([1,2,3,4])
plt.show()
```

```python
#라인 그래프 x축, y축 값 지정

plt.plot([1,2,3,4],[1,4,2,8])
plt.show()
```

```python
x = [1,2,3,4]
y = [1,3,5,9]

plt.plot(x,y,"--r")
plt.plot(x, [4,2,6,4], "--y") # oy = 점으로 찍고 yellow
plt.xlabel("class")
plt.ylabel("count")
plt.title("Title")


plt.show()
```

```python
#subplot - 내 답

x1 = [1,3,4,5]
y1 = [2,5,4,8]
x2 = [7,5,9,8]
y2 = [7,5,3,2]

plt.subplot(2, 2, 1)
plt.plot(x1, y1)

plt.subplot(2, 2, 2)
plt.plot(x2, y2)

plt.subplot(2, 2, 3)
plt.plot(x1, y1)

plt.subplot(2, 2, 4)
plt.plot(x2, y2)
```

```python
#교수님 답
plt.subplot(2,2,1)
plt.plot([1,4,5,6])
plt.subplot(2,2,2)
plt.plot([1,4,5,6])
plt.subplot(2,2,3)
plt.plot([1,4,5,6])
plt.subplot(2,2,4)
plt.bar([1,4,5,6],[1,5,3,4])
plt.show()

```

```python
#막대 그래프 그리기

x = [1,2,3,4]
y = [10,15,21,25]
plt.bar(x,y,color=["r", "y","b","g"], width=0.8)
plt.xticks(x, ["a","b","c","d"])
plt.show()
```

```python
#산점도 그래프
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
# scatter plot에 옵션 추가

color = np.random.rand(50)
area = (50 * np.random.rand(50)) **2 
plt.scatter(x,  y, s = area, c= color, alpha= 0.7)
plt.show()
```

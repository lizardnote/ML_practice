```python
df = pd.read_csv("https://raw.githubusercontent.com/YoungJinBD/data/main/exam/9_1_3.csv")
df.head()
```

```python
df.isna().sum()
```

```python
# 평균만족도 결측치는 컬럼의 전체 평균으로 대체
df['평균만족도'] = df['평균만족도'].fillna(df['평균만족도'].mean())

#근속연수 결측치는 각 부서와 등급별 평균 근속연수로 대체
mean_ten = np.floor(df.groupby(['부서','등급'])['근속연수'].mean()).reset_index().rename(columns = {'근속연수' : '평균근속연수'})
```

```python
mean_ten
```

```python
df = df.merge(mean_ten, on = ['부서','등급'],how = 'left')
df['근속연수'] = np.where(df['근속연수'].isnull(), df['평균근속연수'], df['근속연수'])
df.drop(columns = ['평균근속연수'], inplace = True)
```

```python
df
```

```python
cond1 = (df['부서'] == 'HR') & (df['등급']== 'A')
a = df[cond1]['근속연수'].mean()
```

```python
cond2 = (df['부서'] == 'Sales') & (df['등급']== 'B')
b = df[cond2]['교육참가횟수'].mean()
```

```python
a+b
```

```python
a,b
```

##6-1 유형(2)

```python
df = pd.read_csv("https://raw.githubusercontent.com/YoungJinBD/data/main/exam/6_1_2.csv")
df.head()
```

```python
df['Height_m'] = df['Height_cm'] / 100
df['BMI'] = np.round( df['Weight_kg'] / (df['Height_m'] ** 2) , 1)
```

```python
df.head(2)
```

```python
#직접 할당하기 loc 이용
df1 = df.copy()
df1['category'] = 'unknown'
df1.loc[ df['BMI'] >= 25, 'category'] = '비만'
df1.loc[ (df['BMI'] < 25) & (df['BMI'] >= 23), 'category'] = '과체중'
df1.loc[ (df['BMI'] < 23) & (df['BMI'] >= 18.5), 'category'] = '정상'
df1.loc[ df['BMI'] < 18.5, 'category'] = '저체중'
```

```python
df1.head()
```

```python
df2 = df.copy()
cond = [
    (df2['BMI'] >= 25),
    (df2['BMI'] < 25) & (df2['BMI'] >= 23),
    (df2['BMI'] < 23) & (df2['BMI'] >= 18.5),
    (df2['BMI'] < 18.5)
]

cate = ['비만', '과체중', '정상', '저체중']
df2['category'] = np.select(cond, cate, default='unknown')

df2.head(2)
```

```python
#pd.cut() 사용하기

df3 = df.copy()

bins = [ -float('inf'), 18.5, 23, 25, float('inf') ]
labels = ['저체중', '정상' ,'과제중', '비만']

df3['category'] = pd.cut( df3['BMI'], bins = bins, labels = labels, right = False)
df3.head()
```

```python
df1['category'].value_counts()
```

```python
df2['category'].value_counts()
```

```python
df3['category'].value_counts()
```

```python
df1.groupby('category').count()
```
